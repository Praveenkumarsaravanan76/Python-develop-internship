```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/internships/Analyse data/Data")
print("First 5 rows:")
print(df.head())
print("\nData Summary:")
print(df.describe())
average_rent = df['Rent'].mean()
print(f"\nAverage Rent: â‚¹{average_rent:.2f}")
plt.figure(figsize=(8, 5))
df['Location'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Number of Properties by Location")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 5))
plt.scatter(df['Size'], df['Rent'], color='green', alpha=0.6)
plt.title('House Size vs Rent')
plt.xlabel('Size (sqft)')
plt.ylabel('Rent (â‚¹)')
plt.grid(True)
plt.tight_layout()
plt.show()
plt.figure(figsize=(6, 4))
correlation = df[['Size', 'Bedroom', 'Rent']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
